```python
# reading in the model


def read_in_model():
    # model = load_model("new_run.keras")
    # model.summary()
    model = GenomeMLP()
    
    model(tf.keras.Input(shape=( hp.num_features, )))

    model.load_weights('weights/best_weights.h5')
    
    model.summary()

    model.compile(
        optimizer = model.optimizer, 
        loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False), 
        metrics = [["sparse_categorical_accuracy"]]
    )

    print(model.architecture.outputs)
    return model 

model = read_in_model()
```

```python
# reshape the shap vals
feature_names = X_train.var.index.tolist()
copy_shap_vals = np.squeeze(aggregate_shap)

print(copy_shap_vals.shape)

print(aggregate_shap_sample.shape)

# # convert into an explanation
# shap_vals = shap.Explanation(shap_vals)
# print(shap_vals.values)

print(len(feature_names))

# getting the predictions 

preds = model.predict(aggregate_shap_sample)
preds = np.argmax(tf.nn.softmax(preds, axis = 1), axis = 1)
# we isolate the features of the predicted class to try to see the maximumly impactful features 
# on the prediction. the shap values of the features for the other classes are quite minimal in 
# comparison to the predicted class

new_shap_vals = []
for i, pred in enumerate(preds):
    new_shap_vals.append(copy_shap_vals[i][:, pred])

new_shap_vals = np.array(new_shap_vals)

explanation = shap.Explanation(new_shap_vals, feature_names= feature_names)



shap.plots.bar(explanation, max_display = 10, show = False)
plt.title(f'Aggregate Shap plot values')
plt.show()
```

```python

# Extract the mean absolute SHAP values for each feature
mean_abs_shap_vals = np.mean(np.abs(explanation.values), axis=0)

# Get the indices of the top 110 features
top_100_indices = np.argsort(mean_abs_shap_vals)[-101:][::-1]

# Extract the feature names for the top 110 features
top_100_features = [explanation.feature_names[i] for i in top_100_indices]

print(top_100_features)
```
